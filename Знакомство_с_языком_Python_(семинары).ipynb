```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
```

```python
data = pd.DataFrame({'whoAmI': lst})
data.head()
```

```python
# для проверки
pd.get_dummies(data['whoAmI']).head()
```

```python
# выбираем уникальные значения датафрейма
values_to_headers = data['whoAmI'].unique()
values_to_headers
```

```python
result = pd.DataFrame()
for i in values_to_headers:
    result[i] = data['whoAmI']
    result.loc[result[i]!=i, i] = 0
    result.loc[result[i]==i, i] = 1
```
